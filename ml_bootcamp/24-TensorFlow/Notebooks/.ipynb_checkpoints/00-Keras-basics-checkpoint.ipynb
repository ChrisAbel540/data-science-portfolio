```python
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
# use values to return a numpy array which Tensorflow expects, not a dataframe series
# Use capital to denote a feature matrix
```

```python
y = df['price'].values
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop', loss='mse')
#Final layer has a single output as in this example we're predicting one value
#I prefer this method
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

# Work with the test data

```python
model.evaluate(X_test,y_test,verbose=0)
# Loss according to the units specified in the .compile() spec.
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
pred_df =  pd.DataFrame(y_test, columns = ['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions], axis=1)
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
sns.scatterplot(data=pred_df,x='Test True Y',y='Model Predictions')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
np.sqrt(mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions']))
```
